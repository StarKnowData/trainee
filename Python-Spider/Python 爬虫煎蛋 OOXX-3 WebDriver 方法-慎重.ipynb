```python
#coding:utf-8
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
import requests
import re
import os
import time
from bs4 import BeautifulSoup

browser = webdriver.Chrome()
# 设置网站等待时间
wait = WebDriverWait(browser, 3)

# 文件夹，需要加最后一个斜杠
jandanDir = "/Users/lw/WorkSpace/Anaconda/jandan/"

# 获取 URL 页面的 HTML 源文件
def get_html(url):
    print('正在爬取...'+url)
    try:
        # 浏览器打开一个 URL
        browser.get(url)
        # 获取浏览器当前运行的源代码
        # 注意，这里如果是有 JS 会运行的话，会先运行完 JS
        html=browser.page_source
        
        if html:
            return html
    except EOFError:
        return None

def get_page():
    
    # 第一次进入的原始地址
    base_url = 'http://jandan.net/ooxx/' 
    
    print("进入->"+base_url)
    # 解析原始地址
    soup = BeautifulSoup(get_html(base_url), 'lxml')  
    
    # 找到页面中最大页面的元素，注意，在煎蛋网站上标识最大页面的部分
    current_pages=soup.select('.cp-pagenavi .current-comment-page')[0].get_text()[1:-1] #取出当前页面字段
    
    urllist = []
    
    for page in range(1,int(current_pages)+1):
        # 拼出分页地址
        real_url=base_url+'page-'+str(page)+'#comments' 
        urllist.append(real_url)
    print(len(urllist))
    return urllist #返回所有分页地址列表

def mkdir():
    # 判断是否存在目录
    isExists=os.path.exists(jandanDir)
    if not isExists:
        print('创建目录')
        os.makedirs(jandanDir) #创建目录
        os.chdir(jandanDir) #切换到创建的文件夹
        return True
    else:
        print('目录已存在,即将保存！')
        return False

def get_pic():
    for url in get_page():
        mkdir()
        # 获取页面
        html=get_html(url)
        
        
        soup=BeautifulSoup(html,'lxml')
        allimgs=soup.select('div .text p img')
        allimgs=re.findall('src="(.*?)"',str(allimgs))
        
        
        download(allimgs)
    print("下载完毕！")

# 正式下载图片
def download(allimgs):
    for img_url in allimgs:
        filename = jandanDir+img_url.split('/')[-1]
        # 写入文件
        with open(filename,'wb+') as f:
            try:
                f.write(requests.get(img_url).content)
                print('成功下载图片',filename)
            except:
                print('下载失败',filename)

def main():
    # 计算全部下载完需要多长时间
    t = time.time()
    get_pic()
    print(time.time() - t)

if __name__ == '__main__':
    main()

```
